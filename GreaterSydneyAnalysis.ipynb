# Greater Sydney Analysis – SA4 Blacktown

## 0 Setup

```python
# Python imports
import time, requests
import pandas as pd
import geopandas as gpd
from sqlalchemy import create_engine, inspect
import matplotlib.pyplot as plt


# Database connection
DB_URL = "postgresql+psycopg2://root:pg@localhost:5432/gsa"
engine = create_engine(DB_URL, future=True)
inspector = inspect(engine)

with engine.begin() as conn:
    conn.exec_driver_sql(
        """
        DROP SCHEMA IF EXISTS analysis CASCADE;
        DROP SCHEMA IF EXISTS raw CASCADE;
        CREATE SCHEMA raw;
        CREATE SCHEMA analysis;
    """
    )
```

## TASK 1
### 1.1 Load raw data

```python
# Paths to raw files
SHAPEFILE = "./data_raw/SA2_2021_AUST_GDA2020/SA2_2021_AUST_GDA2020.shp"
BUSINESSES_CSV = "./data_raw/Businesses.csv"
POP_CSV = "./data_raw/Population.csv"
INCOME_CSV = "./data_raw/Income.csv"
STOPS_TXT = "./data_raw/Stops.txt"
PRIMARY_SHP = "./data_raw/catchments/catchments_primary.shp"
SECONDARY_SHP = "./data_raw/catchments/catchments_secondary.shp"

# 1.1 SA2 shapefile
gdf_sa2 = (
    gpd.read_file(SHAPEFILE)
    .to_crs(7844)
    .rename(
        columns={
            "SA2_CODE21": "sa2_code21",
            "SA2_NAME21": "sa2_name21",
            "SA4_NAME21": "sa4_name21",
        }
    )
)
# Save the GeoDataFrame to PostGIS under schema "raw", replacing if exists
gdf_sa2.to_postgis("sa2_all", engine, schema="raw", if_exists="replace", index=False)
print("Reloaded raw.sa2_all")

# Data Cleaning
# 1.2 Businesses
df_biz = pd.read_csv(BUSINESSES_CSV, dtype=str)
df_biz["sa2_code"] = df_biz["sa2_code"].str.zfill(5)
df_biz = df_biz.dropna(subset=["sa2_code", "total_businesses"])
df_biz["total_businesses"] = df_biz["total_businesses"].astype(int)
df_biz.to_sql("businesses", engine, schema="raw", if_exists="replace", index=False)
print("Loaded raw.businesses (cleaned)")

# 1.3 Population
df_pop = pd.read_csv(POP_CSV, dtype=str)
df_pop["sa2_code"] = df_pop["sa2_code"].str.zfill(5)
df_pop = df_pop.dropna(subset=["total_people"])
df_pop["total_people"] = df_pop["total_people"].astype(int)
df_pop.to_sql("population", engine, schema="raw", if_exists="replace", index=False)
print("Loaded raw.population (cleaned)")

# 1.4 Income
df_inc = pd.read_csv(INCOME_CSV, dtype=str)
df_inc["sa2_code21"] = df_inc["sa2_code21"].str.zfill(5)
# Safely convert median_income to numeric, coercing invalid values to NaN
df_inc["median_income"] = pd.to_numeric(df_inc["median_income"], errors="coerce")
df_inc = df_inc.dropna(subset=["median_income"])
df_inc["median_income"] = df_inc["median_income"].astype(int)
df_inc.to_sql("income", engine, schema="raw", if_exists="replace", index=False)
print("Loaded raw.income (cleaned)")

# 1.5 Stops
stops = pd.read_csv(STOPS_TXT)
# Remove rows with missing coordinates and duplicates
stops = stops.dropna(subset=["stop_lon", "stop_lat"]).drop_duplicates(
    subset=["stop_id"]
)
# Convert to GeoDataFrame and reproject
gdf_stops = gpd.GeoDataFrame(
    stops, geometry=gpd.points_from_xy(stops.stop_lon, stops.stop_lat), crs="EPSG:4326"
).to_crs(7844)
# Save to PostGIS
gdf_stops.to_postgis("stops", engine, schema="raw", if_exists="replace", index=False)
print("Loaded raw.stops")

# 1.6 Catchments
for shp, tbl in [
    (PRIMARY_SHP, "catchments_primary"),
    (SECONDARY_SHP, "catchments_secondary"),
]:
    gdf = gpd.read_file(shp).to_crs(7844)
    gdf = gdf[gdf.geometry.is_valid]  # remove invalid geometries
    gdf.to_postgis(tbl, engine, schema="raw", if_exists="replace", index=False)
    print(f"Loaded raw.{tbl}")

print("All cleaned raw.* tables loaded.")


# 2. Add PKs and FKs in SQL
with engine.begin() as conn:
    conn.exec_driver_sql(
        """
        -- SA2 table primary key
        ALTER TABLE raw.sa2_all
          ADD CONSTRAINT pk_sa2_all PRIMARY KEY (sa2_code21);

        -- Businesses
        ALTER TABLE raw.businesses
          -- No duplicate business entries for the same SA2 and industry
          ADD CONSTRAINT pk_businesses PRIMARY KEY (sa2_code, industry_name),
          ADD CONSTRAINT fk_businesses_sa2
            FOREIGN KEY (sa2_code) REFERENCES raw.sa2_all(sa2_code21);

        -- Population 
        ALTER TABLE raw.population
          ADD CONSTRAINT pk_population PRIMARY KEY (sa2_code),
          ADD CONSTRAINT fk_population_sa2
            FOREIGN KEY (sa2_code) REFERENCES raw.sa2_all(sa2_code21);

        -- Income
        ALTER TABLE raw.income
          ADD CONSTRAINT pk_income PRIMARY KEY (sa2_code21),
          ADD CONSTRAINT fk_income_sa2
            FOREIGN KEY (sa2_code21) REFERENCES raw.sa2_all(sa2_code21);

        -- Stops
        ALTER TABLE raw.stops
          ADD CONSTRAINT pk_stops PRIMARY KEY (stop_id);

        -- Catchments: add surrogate key
        ALTER TABLE raw.catchments_primary
          ADD COLUMN catchment_id SERIAL PRIMARY KEY;
        ALTER TABLE raw.catchments_secondary
          ADD COLUMN catchment_id SERIAL PRIMARY KEY;
    """
    )
print("Declared primary and foreign keys on raw.* tables.")
```

### 1.2 Filter to CHOSEN SA4
Create *_my tables limited to the chosen SA4.

```python
with engine.begin() as conn:
    conn.exec_driver_sql(
        """
      DROP TABLE IF EXISTS raw.sa2_my CASCADE;
      CREATE TABLE raw.sa2_my AS
      SELECT
        sa4_name21,
        sa2_code21,
        sa2_name21,
        geometry
      FROM raw.sa2_all
      WHERE sa4_name21 IN (
        'Sydney - North Sydney and Hornsby',
        'Sydney - Eastern Suburbs',
        'Sydney - Blacktown'
      );
      -- Creates a spatial index for fast spatial joins and lookups.
      CREATE INDEX idx_sa2_my_geom ON raw.sa2_my USING gist(geometry);

      ALTER TABLE raw.sa2_my
        ADD CONSTRAINT pk_sa2_my PRIMARY KEY (sa2_code21);
    """
    )

    # 2. Filter each CSV table down to those SA2 codes
    pk_map = {
        "businesses": "(sa2_code21, industry_name)",
        "population": "(sa2_code21)",
        "income": "(sa2_code21)"
    }

    for name in ["businesses", "population", "income"]:
        sa2_col = "sa2_code21" if name == "income" else "sa2_code"
        select_extra = (
            "" if sa2_col == "sa2_code21" else f", t.{sa2_col}::text AS sa2_code21"
        )
        pk = pk_map[name]

        conn.exec_driver_sql(
            f"""
            DROP TABLE IF EXISTS raw.{name}_my CASCADE;
            CREATE TABLE raw.{name}_my AS
            SELECT t.*{select_extra}
            FROM raw.{name} t
            JOIN raw.sa2_my s
              ON t.{sa2_col}::text = s.sa2_code21;

            ALTER TABLE raw.{name}_my
              ADD CONSTRAINT pk_{name}_my PRIMARY KEY {pk};
            """
        )


    # 3. Spatially filter GTFS stops to those three SA4s
    conn.exec_driver_sql(
        """
        DROP TABLE IF EXISTS raw.stops_my CASCADE;
        CREATE TABLE raw.stops_my AS
        SELECT s.*, a.sa4_name21
        FROM raw.stops s
        JOIN raw.sa2_my a
          ON ST_Contains(a.geometry, s.geometry);

        CREATE INDEX IF NOT EXISTS idx_stops_my_geom
          ON raw.stops_my USING gist (geometry);

        ALTER TABLE raw.stops_my
          ADD CONSTRAINT pk_stops_my PRIMARY KEY (stop_id);
    """
    )

    conn.exec_driver_sql(
    """
    -- PRIMARY catchments, spatially joined to SA2s
    DROP TABLE IF EXISTS raw.catchments_primary_my CASCADE;
    CREATE TABLE raw.catchments_primary_my AS
    SELECT 
        c.*, 
        s.sa2_code21,           -- needed for z_schools
        s.sa4_name21
    FROM raw.catchments_primary c
    JOIN raw.sa2_my s
      ON ST_Intersects(s.geometry, c.geometry);

    CREATE INDEX IF NOT EXISTS idx_catchments_primary_my_geom
      ON raw.catchments_primary_my USING gist (geometry);

    -- Add surrogate key
    ALTER TABLE raw.catchments_primary_my
      ADD COLUMN id SERIAL PRIMARY KEY;


    -- SECONDARY catchments, spatially joined to SA2s
    DROP TABLE IF EXISTS raw.catchments_secondary_my CASCADE;
    CREATE TABLE raw.catchments_secondary_my AS
    SELECT 
        c.*, 
        s.sa2_code21,           -- needed for z_schools (if you use secondary)
        s.sa4_name21
    FROM raw.catchments_secondary c
    JOIN raw.sa2_my s
      ON ST_Intersects(s.geometry, c.geometry);

    CREATE INDEX IF NOT EXISTS idx_catchments_secondary_my_geom
      ON raw.catchments_secondary_my USING gist (geometry);

    -- Add surrogate key
    ALTER TABLE raw.catchments_secondary_my
      ADD COLUMN id SERIAL PRIMARY KEY;
    """
)


print("SA2 slice + CSV + stops + catchments all include the three SA4s in raw.sa2_my")
```

## TASK 2
### 2.1 POI API extraction

```python
LAYER_URL = (
    "https://maps.six.nsw.gov.au/arcgis/rest/services/public/NSW_POI/MapServer/0/query"
)

FIELD_LIST = ",".join(
    [
        "objectid",
        "poigroup",
        "poitype",
        "poiname",
        "poilabel",
        "poialtlabel",
    ]
)


def fetch_poi_bbox(bbox, chunk=1000):
    params = {
        "geometry": ",".join(map(str, bbox)),
        "geometryType": "esriGeometryEnvelope",
        "inSR": 4326,
        "outSR": 4326,
        "spatialRel": "esriSpatialRelIntersects",
        "outFields": FIELD_LIST,
        "returnGeometry": "true",
        "where": "1=1",
        "f": "json",
        "resultRecordCount": chunk,
    }

    records, offset = [], 0
    while True:
        params["resultOffset"] = offset
        for attempt in range(5):
            try:
                resp = requests.get(LAYER_URL, params=params, timeout=15)
                resp.raise_for_status()
                break
            except (requests.exceptions.ConnectionError, requests.exceptions.Timeout):
                print(f"ConnectionError, {attempt+1} retries...")
                time.sleep(1 * (2 ** attempt))
            except Exception as e:
                print(e)
                break
        feats = resp.json().get("features", [])
        records.extend(feats)
        if len(feats) < chunk:
            break
        offset += chunk
        time.sleep(1)
    return records
```

### 2.2 Loop through every Blacktown SA2

```python
from shapely.geometry import Point

# Load SA2 geometries and convert to EPSG:4326
gdf_sa2 = gpd.read_postgis(
    "SELECT sa2_code21, geometry FROM raw.sa2_my", engine, geom_col="geometry"
).to_crs(4326)

POI_GROUP_MAP = {
    1: "Community", 2: "Education", 3: "Recreation", 4: "Transport",
    5: "Utility", 6: "Hydrography", 7: "Landform", 8: "Place", 9: "Industry",
}

# Store POI data
poi_rows = []

total = len(gdf_sa2)
for idx, row in enumerate(gdf_sa2.itertuples(index=False), start=1):
    sa2_code = row.sa2_code21
    sa2_geom = row.geometry
    bbox = sa2_geom.bounds

    print(f"[{idx}/{total}] Fetching POIs for SA2 {sa2_code}")
    features = fetch_poi_bbox(bbox)

    for f in features:
        attr = f["attributes"]
        x, y = f["geometry"]["x"], f["geometry"]["y"]
        poi_point = Point(x, y)

        if not sa2_geom.covers(poi_point):
            continue

        poi_rows.append({
            "sa2_code21": sa2_code,
            "poi_group": POI_GROUP_MAP.get(attr.get("poigroup"), f"Unknown({attr.get('poigroup')})"),
            "poi_type": attr.get("poitype"),
            "name": attr.get("poiname") or attr.get("poilabel") or attr.get("poialtlabel"),
            "longitude": x,
            "latitude": y,
        })

    time.sleep(1)

print(f"Collected {len(poi_rows):,} POIs across all SA2s")

# Convert to GeoDataFrame
gdf_poi = gpd.GeoDataFrame(
    poi_rows,
    geometry=gpd.points_from_xy(
        [r["longitude"] for r in poi_rows],
        [r["latitude"] for r in poi_rows]
    ),
    crs="EPSG:4326"
).to_crs("EPSG:7844")

# Keep only useful columns
gdf_poi = gdf_poi[["sa2_code21", "poi_group", "poi_type", "name", "geometry"]]

# Write to PostGIS
with engine.begin() as conn:
    conn.exec_driver_sql("DROP TABLE IF EXISTS raw.poi_my CASCADE;")

gdf_poi.to_postgis(
    name="poi_my",
    con=engine,
    schema="raw",
    if_exists="replace",
    index=False
)

# Create spatial index
with engine.begin() as conn:
    conn.exec_driver_sql("""
        CREATE INDEX IF NOT EXISTS idx_poi_my_geom
        ON raw.poi_my USING gist (geometry);
    """)

print(f"raw.poi_my table created with {len(gdf_poi):,} records.")
```

### 2.3 Ingest Result Into Postgres

```python
# Convert rows to GeoDataFrame
gdf_poi = gpd.GeoDataFrame(
    poi_rows,
    geometry=gpd.points_from_xy(
        [r["longitude"] for r in poi_rows], [r["latitude"] for r in poi_rows]
    ),
    crs="EPSG:4326",
).to_crs("EPSG:7844")

# Filter columns
gdf_poi = gdf_poi[["sa2_code21", "poi_group", "poi_type", "name", "geometry"]]


with engine.begin() as conn:
    conn.exec_driver_sql("DROP VIEW IF EXISTS analysis.vw_metric_poi CASCADE;")
# Save to PostGIS
gdf_poi.to_postgis(
    name="poi_my", con=engine, schema="raw", if_exists="replace", index=False
)

# Create spatial index
with engine.begin() as conn:
    conn.exec_driver_sql(
        """
        CREATE INDEX IF NOT EXISTS idx_poi_my_geom
        ON raw.poi_my USING gist (geometry);
    """
    )

print(f"raw.poi_my table created with {len(gdf_poi):,} rows")
```

## TASK 3
### 3.1 Metric-level views

```python
with engine.begin() as conn:
    conn.exec_driver_sql(
        """
        -- Businesses per 1 000 residents
        CREATE OR REPLACE VIEW analysis.vw_metric_business AS
        SELECT
            b.sa2_code21,
            SUM(b.total_businesses)::numeric
              / NULLIF(MAX(p.total_people), 0) * 1000 AS biz_per_1k
        FROM raw.businesses_my  AS b
        JOIN raw.population_my  AS p USING (sa2_code21)
        WHERE b.industry_name IN (
            'Health Care & Social Assistance',
            'Retail Trade',
            'Education & Training'
        )
        GROUP BY b.sa2_code21;

        -- Public-transport stops (raw count)
        CREATE OR REPLACE VIEW analysis.vw_metric_stops AS
        SELECT
            s.sa2_code21,
            COUNT(*)::numeric AS stops_raw
        FROM raw.stops_my AS st
        JOIN raw.sa2_my   AS s
          ON ST_Contains(s.geometry, st.geometry)
        GROUP BY s.sa2_code21;

        -- POIs in resource-signalling groups (raw count)
        CREATE OR REPLACE VIEW analysis.vw_metric_poi AS
        SELECT
            p.sa2_code21,
            COUNT(*)::numeric AS poi_raw
        FROM raw.poi_my AS p
        WHERE p.poi_group IN ('Medical','Shopping','Recreation')
        GROUP BY p.sa2_code21;

        -- Schools per 1 000 youth (ages 0-19)
        DROP VIEW IF EXISTS analysis.vw_metric_schools CASCADE;

        CREATE OR REPLACE VIEW analysis.vw_metric_schools AS
        SELECT
            pop.sa2_code21,
            (COALESCE(pr.count_pri,0) + COALESCE(sec.count_sec,0))::numeric
              / NULLIF(
                  COALESCE(pop."0-4_people"::integer,0)
                + COALESCE(pop."5-9_people"::integer,0)
                + COALESCE(pop."10-14_people"::integer,0)
                + COALESCE(pop."15-19_people"::integer,0)
                , 0
              ) * 1000 AS schools_per_1k_youth
        FROM raw.population_my AS pop

        -- count primary catchments via spatial join
        LEFT JOIN (
            SELECT
              s.sa2_code21,
              COUNT(*) AS count_pri
            FROM raw.catchments_primary_my AS c
            JOIN raw.sa2_my              AS s
              ON ST_Intersects(c.geometry, s.geometry)
            GROUP  BY s.sa2_code21
        ) AS pr
          ON pr.sa2_code21 = pop.sa2_code21

        -- count secondary catchments via spatial join
        LEFT JOIN (
            SELECT
              s.sa2_code21,
              COUNT(*) AS count_sec
            FROM raw.catchments_secondary_my AS c
            JOIN raw.sa2_my                AS s
              ON ST_Intersects(c.geometry, s.geometry)
            GROUP  BY s.sa2_code21
        ) AS sec
          ON sec.sa2_code21 = pop.sa2_code21;
        """
    )
```

## 3.2 Merge everything into a single per-SA2 view

```python
with engine.begin() as conn:
    conn.exec_driver_sql(
        """
        DROP VIEW IF EXISTS analysis.vw_sa2_metrics CASCADE;

        CREATE VIEW analysis.vw_sa2_metrics AS
        SELECT
            s.sa2_code21,
            COALESCE(b.biz_per_1k,        0) AS business,
            COALESCE(t.stops_raw,         0) AS stops,
            COALESCE(p.poi_raw,           0) AS poi,
            COALESCE(sch.schools_per_1k_youth, 0) AS schools,
            pop.total_people
        FROM raw.sa2_my               AS s
        LEFT JOIN analysis.vw_metric_business AS b USING (sa2_code21)
        LEFT JOIN analysis.vw_metric_stops    AS t USING (sa2_code21)
        LEFT JOIN analysis.vw_metric_poi      AS p USING (sa2_code21)
        LEFT JOIN analysis.vw_metric_schools  AS sch USING (sa2_code21)
        LEFT JOIN raw.population_my           AS pop USING (sa2_code21)
        WHERE pop.total_people >= 100;   -- drop tiny SA2s
    """
    )
```

## 3.3 Z-scores via window functions

```python
with engine.begin() as conn:
    conn.exec_driver_sql(
        """
        DROP VIEW IF EXISTS analysis.vw_sa2_zscores CASCADE;

        /* Z-scores via window functions, now including schools */
        CREATE OR REPLACE VIEW analysis.vw_sa2_zscores AS
        SELECT
            sa2_code21,
            business,
            stops,
            poi,
            schools,
            (business - AVG(business) OVER ())
              / NULLIF(STDDEV_POP(business) OVER (), 0) AS z_business,
            (stops    - AVG(stops)    OVER ())
              / NULLIF(STDDEV_POP(stops)    OVER (), 0) AS z_stops,
            (poi      - AVG(poi)      OVER ())
              / NULLIF(STDDEV_POP(poi)      OVER (), 0) AS z_poi,
            (schools  - AVG(schools)  OVER ())
              / NULLIF(STDDEV_POP(schools)  OVER (), 0) AS z_schools
        FROM analysis.vw_sa2_metrics;
    """
    )
```

## 3.4 Final scores table in `analysis` schema

```python
with engine.begin() as conn:
    conn.exec_driver_sql(
        """
        -- Final scores table: z_total
        DROP TABLE IF EXISTS analysis.sa2_scores;
        CREATE TABLE analysis.sa2_scores AS
        SELECT
            sa2_code21,
            sa4_name21,
            z_business,
            z_stops,
            z_poi,
            z_schools,
            (z_business + z_stops + z_poi + z_schools) AS z_total,
            1 / (1 + EXP(-(z_business + z_stops + z_poi + z_schools))) AS score
        FROM analysis.vw_sa2_zscores;

        ALTER TABLE analysis.sa2_scores
            ADD CONSTRAINT sa2_scores_pk PRIMARY KEY (sa2_code21);

        CREATE INDEX IF NOT EXISTS idx_sa2_scores_score
            ON analysis.sa2_scores (score);
    """
    )
print("analysis.sa2_scores created")
```

```python
sql = """
SELECT
  s.sa2_code21,
  s.sa4_name21,
  s.z_business,
  s.z_stops,
  s.z_poi,
  s.z_schools,
  s.z_total,
  s.score,
  i.median_income
FROM analysis.sa2_scores AS s
LEFT JOIN raw.income_my AS i
  ON s.sa2_code21 = i.sa2_code21
;
"""

df = pd.read_sql_query(sql, con=engine)


# Aggregate SA2 to SA4 summary
sa4_summary = (
    df.groupby("sa4_name21")["score"]
    .agg(["mean", "median", "min", "max", "count"])
    .rename(
        columns={
            "mean": "avg_score",
            "median": "med_score",
            "min": "min_score",
            "max": "max_score",
            "count": "n_sa2",
        }
    )
    .reset_index()
)

print(sa4_summary)
```

```python
sql = """
SELECT a.sa2_code21,
       a.sa2_name21,
       a.geometry,
       s.sa4_name21,
       s.score
FROM raw.sa2_my AS a
JOIN analysis.sa2_scores AS s
  ON a.sa2_code21 = s.sa2_code21
"""
sa2_gdf = gpd.read_postgis(sql, con=engine, geom_col="geometry")

# Boxplot of SA2 score distribution within each SA4
fig, ax = plt.subplots(figsize=(8, 5))
sa2_gdf.boxplot(column="score", by="sa4_name21", ax=ax)
ax.set_title("Score Distribution by SA4")
ax.set_xlabel("SA4")
ax.set_ylabel("Score")
plt.suptitle("")  # removes the automatic “Boxplot grouped by …” subtitle
plt.xticks(rotation=30, ha="right")
plt.tight_layout()
plt.show()

# Choropleth map of SA2 scores
plt.figure(figsize=(8, 8))
sa2_gdf.plot(column="score", legend=True)
plt.title("Choropleth of SA2 Scores")
plt.axis("off")
plt.tight_layout()
plt.show()
```
